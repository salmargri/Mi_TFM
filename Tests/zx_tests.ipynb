```python
import pyzx as zx

circuito = zx.Circuit(2)


circuito.add_gate("H", 0)       
circuito.add_gate("CNOT", 0, 1)


circuito.add_gate("Z", 0)
circuito.add_gate("NOT", 0) 


circuito.add_gate("CNOT", 0, 1)
circuito.add_gate("H", 0)

grafo = circuito.to_graph()

zx.draw(grafo)
```

```python
import pyzx as zx

circuito = zx.Circuit(4)

circuito.add_gate("H", 2)
circuito.add_gate("CNOT", 2, 3)


circuito.add_gate("CNOT", 1, 2) 
circuito.add_gate("CZ", 0, 2)

circuito.add_gate("CNOT", 2, 3)
circuito.add_gate("H", 2)

grafo = circuito.to_graph()
zx.draw(grafo)

zx.full_reduce(grafo)
zx.draw(grafo)
```

```python
import pyzx as zx

circ = zx.Circuit(3)
circ.add_gate("H", 1)
circ.add_gate("CNOT", 1, 2)
circ.add_gate("CNOT", 0, 1)
circ.add_gate("H", 0)
circ.add_gate("CNOT", 1, 2)
circ.add_gate("CZ", 0, 2)

g = circ.to_graph()

salida_0 = g.outputs()[0]
salida_1 = g.outputs()[1]


g.set_type(salida_0, zx.VertexType.Z)
g.set_type(salida_1, zx.VertexType.Z)

zx.draw(g)

zx.full_reduce(g)

print("Grafo de Teleportación Real:")
zx.draw(g)
```
